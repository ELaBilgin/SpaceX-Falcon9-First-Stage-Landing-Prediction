```python
df = df.drop(['FlightNumber','BoosterVersion','Longitude','Latitude','Date'], axis = 1)
```

```python
df['GridFins'] = df['GridFins'].apply(int)
df['Reused'] = df['Reused'].apply(int)
df['Legs'] = df['Legs'].apply(int)
```

```python
df_dummy = pd.get_dummies(df, columns=['Orbit','LaunchSite','Outcome', 'LandingPad', 'Serial'], dtype = int)
df_dummy
```

```python
df = pd.concat([df, df_dummy], axis = 1)
```
